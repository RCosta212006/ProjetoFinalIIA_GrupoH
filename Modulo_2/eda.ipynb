```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Configuração visual
sns.set(style="whitegrid", palette="deep")
plt.rcParams["figure.figsize"] = (10, 6)

print("A carregar os dados...")

# Pasta actual do notebook: Modulo_2
pasta_actual = Path.cwd()

# Pasta principal: ProjetoFinalIIA_GrupoH
pasta_projeto = pasta_actual.parent

# CSV fora da pasta Modulo_2
caminho_csv = pasta_projeto / "processed_lisboa_porto_air_quality.csv"

df = pd.read_csv(caminho_csv, sep=';')

print("\n==============================")
print("DIMENSÕES DO DATASET")
print("==============================")
print(f"Linhas: {df.shape[0]}")
print(f"Colunas: {df.shape[1]}")

print("\n==============================")
print("TIPOS DE DADOS")
print("==============================")
display(df.dtypes)

print("\n==============================")
print("ESTATÍSTICAS DESCRITIVAS")
print("==============================")
display(df.describe())

print("\n==============================")
print("VALORES EM FALTA")
print("==============================")
missing = df.isnull().sum()

missing_df = pd.DataFrame({
    'Coluna': missing.index,
    'Valores em Falta': missing.values,
    'Percentagem (%)': ((missing.values / len(df)) * 100).round(2)
})

display(missing_df)

# =========================================================
# COLUNAS NUMÉRICAS
# =========================================================

colunas_numericas = df.select_dtypes(
    include=['float64', 'int64']
).columns

# =========================================================
# HISTOGRAMAS MELHORADOS
# =========================================================

print("\n==============================")
print("DISTRIBUIÇÃO DAS VARIÁVEIS")
print("==============================")

df[colunas_numericas].hist(
    figsize=(18, 14),
    bins=25,
    edgecolor='black'
)

plt.suptitle(
    "Distribuição das Variáveis Numéricas",
    fontsize=18,
    fontweight='bold'
)

plt.tight_layout()
plt.show()

# =========================================================
# BOXPLOTS PARA OUTLIERS
# =========================================================

print("\n==============================")
print("DETECÇÃO DE OUTLIERS")
print("==============================")

num_cols = len(colunas_numericas)

fig, axes = plt.subplots(
    nrows=(num_cols // 2) + (num_cols % 2),
    ncols=2,
    figsize=(16, num_cols * 3)
)

axes = axes.flatten()

for i, col in enumerate(colunas_numericas):

    sns.boxplot(
        x=df[col],
        ax=axes[i]
    )

    axes[i].set_title(f'Boxplot - {col}')

# Remove gráficos vazios
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()

# =========================================================
# DETECÇÃO AUTOMÁTICA DE OUTLIERS (IQR)
# =========================================================

print("\n==============================")
print("OUTLIERS DETECTADOS (IQR)")
print("==============================")

for col in colunas_numericas:

    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR

    outliers = df[
        (df[col] < limite_inferior) |
        (df[col] > limite_superior)
    ]

    percentagem = (len(outliers) / len(df)) * 100

    print(f"\n{col}")
    print(f"Outliers detectados: {len(outliers)}")
    print(f"Percentagem: {percentagem:.2f}%")

# =========================================================
# MATRIZ DE CORRELAÇÃO PROFISSIONAL
# =========================================================

print("\n==============================")
print("MATRIZ DE CORRELAÇÕES")
print("==============================")

corr = df[colunas_numericas].corr()

plt.figure(figsize=(14, 10))

mask = np.triu(np.ones_like(corr, dtype=bool))

sns.heatmap(
    corr,
    mask=mask,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    square=True
)

plt.title(
    "Mapa de Correlação entre Variáveis",
    fontsize=18,
    fontweight='bold'
)

plt.tight_layout()
plt.show()


print("\nEDA concluída com sucesso.")
```
